## Part 1: Data Exploration and Preparation

```python
# Import libraries
import pandas as pd
```

```python
# Load the Excel file
df = pd.read_excel("data/Data Insights - Synthetic Dataset.xlsx")

# Create a copy of original data

df_1 = df.copy()

# Display the first few rows of the dataframe
df_1.head()
```

```python
# Check column data types
df_1.dtypes
```

```python
# Check statistics on data
df_1.describe()
```

By checking the data types, we can see there are some columns that have incorrect data types. AdmissionTime and SeparationTime should be in datetime format and PharmacyCharge should be a float (for some reason its appearing as an object). When handling clinical data, patient details such as DOB and area codes are considered sensitive information and if there is no purpose to include these columns in the data, these columns should be removed.

Using .describe(), we can observe that there are 30000 records in the data set. Columns that have counts of less than 30000 records mean that there are NaN values within the columns. Depending on the context of the column, these NaN values will be handled differently.

We can also see that two columns CCU_Charges and ICU_Charge have different column name formatting. Most charges column names are in CamelCase and so we should standardise the column names to have the same formatting. There is also an inconsistency between using singular and plural terms for Charge.

The PharmacyCharge values are in e+. This column should be checked for possible errors as there is a low possibility that the pharmacy charges would be a number that large. For the purpose of this technical, we will drop this column. 

```python
# Drop columns
df_1 = df_1.drop(columns=['DateOfBirth', 'Postcode', 'PharmacyCharge'])
```

```python
# Rename columns
df_1.rename(columns={'CCU_Charges': 'CCUCharges', 
                     'AccommodationCharge': 'AccommodationCharges', 
                     'ICU_Charge': 'ICUCharges',
                     'TheatreCharge': 'TheatreCharges',
                     'ProsthesisCharge': 'ProsthesisCharges',
                     }, inplace=True)
```

```python
# Change AdmissionTime datetime format
df_1['AdmissionTime'] = pd.to_datetime(df_1['AdmissionTime'], format='%H:%M:%S')

# Change SeparationTime datetime format
df_1['SeparationTime'] = pd.to_datetime(df_1['SeparationTime'], format='%H:%M:%S')
```

```python
df_1.dtypes
```

```python
# Calculate the sum of total charges and create a new column TotalCharges
df_1['TotalCharges'] = df_1[['AccommodationCharges', 'CCUCharges', 'ICUCharges', 
                          'TheatreCharges', 'ProsthesisCharges', 
                          'OtherCharges', 'BundledCharges']].sum(axis=1)
```

## Part 2: Data Analysis and Visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Group by 'caretype' and sum the 'totalcharges' column
grouped_df = df_1.groupby('AR-DRG')['TotalCharges'].sum().reset_index()
grouped_df_top10 = grouped_df.sort_values(by='TotalCharges', ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Plot a bar chart using Matplotlib
plt.bar(grouped_df_top10['AR-DRG'], grouped_df_top10['TotalCharges'])

# Add titles and labels
plt.xlabel('AR-DRG')
plt.ylabel('Total Charges')
plt.title('Total Charges by AR-DRG')

# Show the plot
plt.show()
```

From the barchart above, we know that DRG002, DRG003 and DGR001 are the top 3 largest charges. We will now do a drill down on these DRG groups.

```python
# Make a trend graph on DRG trend changes through time 1/8/2022 to 31/7/2024

# Extract month from admission date
df_1['AdmissionMonth'] = df_1['AdmissionDate'].dt.to_period('M')  # Period format for grouping by month

# Filter the data for DRGs001, 002, 003
filtered_df = df_1[df_1['AR-DRG'].isin(['DRG001', 'DRG002', 'DRG003'])].reset_index(drop=True)

# Group by admission month and DRG, then count the occurrences
grouped_df_1 = filtered_df.groupby(['AdmissionMonth', 'AR-DRG']).size().unstack(fill_value=0)

# Resize the figure
fig, ax = plt.subplots(figsize=(15, 6))

# Plot the line graph
grouped_df_1.plot(kind='line', marker='o', ax=ax)

# Add titles and labels
plt.xlabel('Admission Month/Year')
plt.ylabel('Count of DRGs')
plt.title('DRGs (001, 002, 003) Trend by Admission Month')

# Get all unique admission months from the grouped DataFrame and show all on x label
months = df_1['AdmissionMonth'].unique()

# Set the x-ticks and rotate them by 30 degrees
plt.xticks(ticks=months, labels=months, rotation=30)

# Show the plot
plt.show()
```

### Querying with SQL

```python
import pandasql as psql
```

**Error handling**

AdmissionMonth in our data in currently typed as Period, therefore we will convert the datatype to datetime formate, as SQL does not have period data types and will give a ProgrammingError.

```python
# Convert AdmissionMonth from string to datetime
df_1['AdmissionMonth'] = df_1['AdmissionMonth'].dt.to_timestamp()

# Reformat from timestamp to MM-YYYY
df_1['AdmissionMonth'] = df_1['AdmissionMonth'].dt.strftime('%m-%Y')
```

```python
# There are 23 months of data in this dataset
query = "SELECT AdmissionMonth AS Period, COUNT(episode_id) AS total_admissions, COUNT(episode_id)/23 as average_admission FROM df_1 GROUP BY AdmissionMonth ORDER BY SUBSTR(Period, 4, 4), SUBSTR(Period, 1, 2)"

# Execute the query
result = psql.sqldf(query, locals())
```

```python
# Q5
query2 = "WITH RankedData AS (SELECT Sex, PrincipalDiagnosis, TotalCharges, ROW_NUMBER() OVER (PARTITION BY Sex, PrincipalDiagnosis ORDER BY TotalCharges) AS row, COUNT(*) OVER (PARTITION BY PrincipalDiagnosis, Sex) AS cnt FROM df_1), Percentile AS (SELECT Sex, PrincipalDiagnosis, MAX(CASE WHEN cnt = 1 THEN TotalCharges WHEN row = ROUND(cnt * 0.25) THEN TotalCharges END) AS percentile_25, MAX(CASE WHEN cnt = 1 THEN TotalCharges WHEN row = ROUND(cnt * 0.50) THEN TotalCharges END) AS percentile_50, MAX(CASE WHEN cnt = 1 THEN TotalCharges WHEN row = ROUND(cnt * 0.75) THEN TotalCharges END) AS percentile_75 FROM RankedData GROUP BY PrincipalDiagnosis, Sex) SELECT Sex, PrincipalDiagnosis, percentile_25, percentile_50, percentile_75 FROM Percentile ORDER BY PrincipalDiagnosis, Sex"
result2 = psql.sqldf(query2, locals())
result2
```
